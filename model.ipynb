```python
def _create_frequency_matrix(sentences):
    """
    we create a dictionary for the word frequency table.
    For this, we should only use the words that are not part of the stopWords array.
    Removing stop words and making frequency table
    Stemmer - an algorithm to bring words to its root word.
    :rtype: dict
    """
    frequency_matrix = {}
    stopWords = set(stop_words)
    #ps = PorterStemmer()

    for sent in sentences:
        freq_table = {}
        words = word_tokenize(sent)
        for word in words:
            word = word.lower()
            flag = 0
            inde = []
            word, inde = kannada_root(word, inde)
            if word in stopWords:
                continue

            if word in freq_table:
                freq_table[word] += 1
            else:
                freq_table[word] = 1

        frequency_matrix[sent[:15]] = freq_table

    return frequency_matrix
```

```python
def _create_documents_per_words(freq_matrix):
    word_per_doc_table = {}

    for sent, f_table in freq_matrix.items():
        for word, count in f_table.items():
            if word in word_per_doc_table:
                word_per_doc_table[word] += 1
            else:
                word_per_doc_table[word] = 1

    return word_per_doc_table
```

```python
def _create_tf_matrix(freq_matrix, doc_table):
    tf_matrix = {}
    
    count_words_in_document = len(doc_table)
        
    for sent, f_table in freq_matrix.items():
        tf_table = {}

        for word, count in f_table.items():
            tf_table[word] = doc_table[word] / count_words_in_document

        tf_matrix[sent] = tf_table

    return tf_matrix
```

```python
def _create_idf_matrix(freq_matrix, count_doc_per_words, total_documents):
    idf_matrix = {}

    alldocs = pd.read_csv('train.csv',usecols = ['headline'])
    wordcount = 1
    
    for sent, f_table in freq_matrix.items():
        idf_table = {}

        for word in f_table.keys():
            
            for document in alldocs:
                if word in document:
                    wordcount+=1
                    
            idf_table[word] = math.log10(total_documents / float(wordcount))

        idf_matrix[sent] = idf_table

    return idf_matrix
```

```python
def _create_tf_idf_matrix(tf_matrix, idf_matrix):
    tf_idf_matrix = {}

    for (sent1, f_table1), (sent2, f_table2) in zip(tf_matrix.items(), idf_matrix.items()):

        tf_idf_table = {}

        for (word1, value1), (word2, value2) in zip(f_table1.items(),
                                                    f_table2.items()):  # here, keys are the same in both the table
            tf_idf_table[word1] = float(value1 * value2)

        tf_idf_matrix[sent1] = tf_idf_table

    return tf_idf_matrix
```

```python
def _create_gss_matrix(freq_matrix,category,total_documents):
    
    #print(category)
    gss_matrix = {}
    
    alldocs = pd.read_csv('train.csv', usecols = ['headline','label'])
    doccount = total_documents
    #print(alldocs.head())
    
    for sent,f_table in freq_matrix.items(): 
        gss_table = {}
        a=0
        b=0
        c=0
        d=0
        for word in f_table.keys():
            for i in range(len(alldocs['headline'])):
                if word in alldocs['headline'][i]:
                    if alldocs['label'][i]==category:
                        a+=1
                    else:
                        b+=1
                else:
                    if alldocs['label'][i]==category:
                        c+=1
                    else:
                        d+=1
            
            #print(a," ",b," ",c," ",d)
            p1 = float(a/doccount)
            p2 = float(d/doccount)
            p3 = float(c/doccount)
            p4 = float(b/doccount)
            gss_table[word] = float(p1*p2) - float(p3*p4)
            #print(word," ",gss_table[word])
            
            gss_matrix[sent] = gss_table
    return gss_matrix
```

```python
def _create_tf_gss_matrix(tf_matrix,gss_matrix):
    tf_gss_matrix = {}
    
    for (sent1, f_table1), (sent2, f_table2) in zip(tf_matrix.items(), gss_matrix.items()):

        tf_gss_table = {}

        for (word1, value1), (word2, value2) in zip(f_table1.items(),
                                                    f_table2.items()):  # here, keys are the same in both the table
            tf_gss_table[word1] = float(value1 * value2)

        tf_gss_matrix[sent1] = tf_gss_table
        
    return tf_gss_matrix
```

```python
def _score_sentences(tf_idf_matrix,tf_gss_matrix) -> dict:
    """
    score a sentence by its word's TF-IDF, TF_GSS, Position Of sentence
    Basic algorithm: adding the TF frequency of every non-stop word in a sentence divided by total no of words in a sentence.
    :rtype: dict
    """
    alpha  = 0.75
    beta = 0.85
    gamma = 0.15
    
    sentenceValue1 = {}

    for sent, f_table in tf_idf_matrix.items():
        total_score_per_sentence = 0

        count_words_in_sentence = len(f_table)
        for word, score in f_table.items():
            total_score_per_sentence += score

        sentenceValue1[sent] = total_score_per_sentence / count_words_in_sentence
    
    sentenceValue2 = {}
    
    for sent, f_table in tf_gss_matrix.items():
        total_score_per_sentence = 0

        count_words_in_sentence = len(f_table)
        for word, score in f_table.items():
            total_score_per_sentence += score

        sentenceValue2[sent] = total_score_per_sentence / count_words_in_sentence
        
    sentenceValue3 = {}
    doc_length = len(sentenceValue1)
    
    pos = 0
    for sent in sentenceValue2:
        sentenceValue3[sent] = abs(math.cos(pos*2*math.pi/doc_length))
        pos+=1
        
    for sent in sentenceValue2:
        if sent in sentenceValue1:
            sentenceValue1[sent] = (alpha * sentenceValue1[sent]) + (beta* sentenceValue2[sent]) + (gamma * sentenceValue3[sent])
    else:
        pass
    
    return sentenceValue1
```

```python
def _find_average_score(sentenceValue, num_sent) -> int:
    """
    Find the average score from the sentence value dictionary
    :rtype: int
    """
    '''sumValues = 0
    for entry in sentenceValue:
        sumValues += sentenceValue[entry]

    # Average value of a sentence from original summary_text
    average = (sumValues / len(sentenceValue))
    '''
    
    sort_sent_value = dict(sorted(sentenceValue.items(), key=lambda x: x[1], reverse=True))
    
    if(num_sent == 0):
        return (list(sort_sent_value.items())[0][1]+1)
    
    
    count=0
    
    for i in sort_sent_value.keys():
        if count==num_sent-1:
            key =  i
            break
        else:
            count+=1
    
    threshold = sort_sent_value[key]
    
    return threshold
```

```python
def _generate_summary(sentences, sentenceValue, threshold):
    sentence_count = 0
    summary = ''

    for sentence in sentences:
        if sentence[:15] in sentenceValue and sentenceValue[sentence[:15]] >= (threshold):
            summary += " " + sentence
            sentence_count += 1
    
    if sentence_count == 0:
        return "NIL"
    return summary
```

```python
def run_summarization(text,category,num_sent):
    """
    :param text: Plain summary_text of long article
    :return: summarized summary_text
    """

    '''
    We already have a sentence tokenizer, so we just need 
    to run the sent_tokenize() method to create the array of sentences.
    '''
    # 1 Sentence Tokenize
    sentences = sent_tokenize(text)
    #sentences = clean_text(raw_sentences)
    total_length = len(sentences)
    total_documents = 15
    #print(sentences)

    # 2 Create the Frequency matrix of the words in each sentence.
    freq_matrix = _create_frequency_matrix(sentences)
    #print(freq_matrix)

    # 4 creating table for documents per words
    count_doc_per_words = _create_documents_per_words(freq_matrix)
    #print(count_doc_per_words)

    '''
    Term frequency (TF) is how often a word appears in a document, divided by how many words are there in a document.
    '''
    # 3 Calculate TermFrequency and generate a matrix
    tf_matrix = _create_tf_matrix(freq_matrix,count_doc_per_words)
    #print(tf_matrix)

    
    '''
    Inverse document frequency (IDF) is how unique or rare a word is.
    '''
    # 5 Calculate IDF and generate a matrix
    idf_matrix = _create_idf_matrix(freq_matrix, count_doc_per_words, total_documents)
    #print(idf_matrix)

    # 6 Calculate TF-IDF and generate a matrix
    tf_idf_matrix = _create_tf_idf_matrix(tf_matrix, idf_matrix)
    #print(tf_idf_matrix)
    
    # 5 Calculate GSS and generate a matrix
    gss_matrix = _create_gss_matrix(freq_matrix, category, total_documents)
    #print(gss_matrix)

    # 6 Calculate TF-GSS and generate a matrix
    tf_gss_matrix = _create_tf_gss_matrix(tf_matrix, gss_matrix)
    #print(tf_gss_matrix)

    # 7 Important Algorithm: score the sentences
    sentence_scores = _score_sentences(tf_idf_matrix,tf_gss_matrix)
    #print(sentence_scores)

    # 8 Find the threshold
    threshold = _find_average_score(sentence_scores,num_sent)
    #print(threshold)

    # 9 Important Algorithm: Generate the summary
    summary = _generate_summary(sentences, sentence_scores, threshold)
    return summary
    
```

```python
text_str = '''ಭಗವಾನ್ ರಾಮನು ಹಿಂದೂ ಧರ್ಮದಲ್ಲಿ ಅತ್ಯಂತ ಗೌರವಾನ್ವಿತ ಮತ್ತು ಪೂಜ್ಯ ದೇವರುಗಳಲ್ಲಿ ಒಬ್ಬ. ಅವರು ವಿಷ್ಣುವಿನ ಏಳನೇ ಅವತಾರ ಎಂದು ನಂಬಲಾಗಿದೆ. ಧಾರ್ಮಿಕ ಪುಸ್ತಕಗಳಲ್ಲಿ ವಿವರಿಸಿದ ನಾಲ್ಕು ಆವರ್ತ ಯುಗಗಳಲ್ಲಿ ತ್ರೇತ ಯುಗದಲ್ಲಿ ಭಗವಾನ್ ರಾಮ ಜನಿಸಿದರು. ಭಗವಾನ್ ರಾಮನನ್ನು ಅತ್ಯಂತ ಆದರ್ಶ ಮನುಷ್ಯ, ಮರ್ಯಾದ ಪುರುಷೋತ್ತಮ್ ಎಂದು ಪರಿಗಣಿಸಲಾಗುತ್ತದೆ, ಅವರು ತಮ್ಮಲ್ಲಿ ಎಲ್ಲಾ ಪರಿಪೂರ್ಣ ಮೌಲ್ಯಗಳು ಮತ್ತು ಸದ್ಗುಣಗಳನ್ನು ಹೊಂದಿದ್ದಾರೆ. ಹಿಂದೂಗಳು ಭಗವಾನ್ ರಾಮನ ಗುಣಗಳನ್ನು ನೋಡುತ್ತಾರೆ. ಅವರು ಎಲ್ಲ ಕಾಲದಲ್ಲೂ ಅತ್ಯಂತ ದೈವಿಕ ವ್ಯಕ್ತಿಯಾಗಿದ್ದಾರೆ. ಅವನು ತನ್ನ ಸುವರ್ಣ ದಿನಗಳಲ್ಲಿ ಮತ್ತು ಕಷ್ಟದ ಅವಧಿಯಲ್ಲಿ ಪರಿಪೂರ್ಣ ಮನುಷ್ಯನಾಗಿ ತನ್ನನ್ನು ತಾನು ಅತ್ಯುತ್ತಮವಾಗಿ ನಡೆಸಿಕೊಂಡನು. ಎಲ್ಲಾ ಸಮಯದಲ್ಲೂ, ಅವರು ಪರಿಚಿತ ಮಾನವನಾಗಿ ಆದರ್ಶಪ್ರಾಯ ಮನೋಭಾವವನ್ನು ಪ್ರದರ್ಶಿಸಿದರು.

 

ಭಗವಾನ್ ರಾಮನು ಭಕ್ತ ಮಗನಾಗಿ
ರಾಜ ದಶರಥನು ರಾಮನನ್ನು ತನ್ನ ಉತ್ತರಾಧಿಕಾರಿಯಾಗಿ ಪಟ್ಟಾಭಿಷೇಕ ಮಾಡಲು ನಿರ್ಧರಿಸಿದಾಗ, ಸಾಮ್ರಾಜ್ಯದ ಪ್ರತಿಯೊಂದು ವಿಷಯವೂ ಸುದ್ದಿಗಳಿಂದ ಸಂತೋಷವಾಯಿತು. ಹೇಗಾದರೂ, ಅವನ ಕಿರೀಟಧಾರಣೆಗೆ ಒಂದು ದಿನ ಮೊದಲು, ಕೈಕೇಯಿಯನ್ನು ತನ್ನ ದುಷ್ಟ ಸೇವಕಿ ಮಂಥಾರ ತನ್ನ ಮಗ ಭರತನ ಭವಿಷ್ಯದ ಅಭದ್ರತೆಯ ಬಗ್ಗೆ ಮೆದುಳು ತೊಳೆಯುತ್ತಾನೆ. ಆರಂಭದಲ್ಲಿ ರಾಮನ ಬಗ್ಗೆ ಸಂತಸಗೊಂಡಿದ್ದ ಕೈಕೇಯಿ ಈಗ ಹೆಚ್ಚು ಅಸೂಯೆ ಪಟ್ಟಿದ್ದಾಳೆ ಮತ್ತು ದಶರಥನು ತನಗೆ ನೀಡಿದ ಎರಡು ವರಗಳನ್ನು, ರಾಮನ ಜನನದ ನಂತರ ಮತ್ತು ಉಳಿದ ಒಡಹುಟ್ಟಿದವರನ್ನು ಪೂರೈಸುವಂತೆ ಕೇಳುತ್ತಾನೆ. ಅವಳು ಕೇಳಿದ ಮೊದಲ ವರವೆಂದರೆ ರಾಮನನ್ನು ಹದಿನಾಲ್ಕು ವರ್ಷಗಳ ಕಾಲ ಕಾಡುಗಳಲ್ಲಿ ಬಹಿಷ್ಕಾರಕ್ಕೆ ಕಳುಹಿಸುವುದು, ಮತ್ತು ಎರಡನೆಯದು ಭರತನನ್ನು ರಾಮನ ಸ್ಥಾನದಲ್ಲಿ ಅಯೋಧ್ಯೆಯ ರಾಜನಾಗಿ ಕಿರೀಟಧಾರಣೆ ಮಾಡುವುದು. ದಶರಥನು ಗೊಂದಲದ ಸ್ಥಿತಿಯಲ್ಲಿದ್ದರೆ, ಅದೇ ಸಮಯದಲ್ಲಿ ಅವನು ತನ್ನ ಮಾತಿನ ರಾಜನೆಂದು ತಿಳಿದುಬಂದನು. ರಾಮ, ತನ್ನ ತಂದೆಯ ಬಿಕ್ಕಟ್ಟನ್ನು ಏಕಕಾಲದಲ್ಲಿ ಪರಿಹರಿಸುತ್ತಾನೆ, ಗಡಿಪಾರು ಅವಧಿಯನ್ನು ಸ್ವಇಚೆಯಿಂದ ಸ್ವೀಕರಿಸುವ ಮೂಲಕ ಮತ್ತು ಕಣ್ಣು ಮಿಟುಕಿಸದೆ ತನ್ನ ಶೀರ್ಷಿಕೆಯನ್ನು ಬಿಟ್ಟುಕೊಡುತ್ತಾನೆ, ಇದರಿಂದಾಗಿ ಅವನ ತಂದೆ ಕೈಕೇಯಿಗೆ ನೀಡಿದ ಭರವಸೆಯನ್ನು ಹಿಂತಿರುಗಿಸಬೇಕಾಗಿಲ್ಲ. ರಾಮ, ನಂತರ ತನ್ನ ವನವಾಸಕ್ಕೆ ಹೊರಡುತ್ತಾನೆ, ಸೀತೆ ಮತ್ತು ಲಕ್ಷ್ಮಣ ಸೇರಿಕೊಂಡನು. ಮಗ ತೊರೆದ ನಂತರ ದಶರಥನಿಗೆ ದುಃಖವನ್ನು ಭರಿಸಲಾಗಲಿಲ್ಲ ಮತ್ತು ನಂತರ ಸಾಯುತ್ತಾನೆ. ರಾಮ, ತನ್ನ ತಂದೆಯ ಮರಣದ ನಂತರವೂ ತನ್ನ ಅಧಿಕಾರಾವಧಿಯು ಮುಗಿಯುವವರೆಗೂ ಅಯೋಧ್ಯೆಗೆ ಹಿಂತಿರುಗಬಾರದೆಂದು ನಿರ್ಧರಿಸಿದನು. ಹೀಗಾಗಿ, ಅವನು ಮರಣಿಸಿದ ನಂತರವೂ ಅವನು ತನ್ನ ತಂದೆಗೆ ನೀಡಿದ ವಾಗ್ದಾನವನ್ನು ಈಡೇರಿಸಬೇಕಾದ ಸಮರ್ಪಣೆಯ ಪ್ರಮಾಣವನ್ನು ನಾವು ನೋಡುತ್ತೇವೆ. ಆದ್ದರಿಂದ, ರಾಮನು ತನ್ನ ಹೆತ್ತವರಿಗಾಗಿ ಏನು ಬೇಕಾದರೂ ತ್ಯಾಗ ಮಾಡುವ ಆದರ್ಶ ಭಕ್ತ ಮಗನನ್ನು ಸಂಕೇತಿಸುತ್ತಾನೆ. ರಾಮನ ನೈಸರ್ಗಿಕ ಆತ್ಮವು ವಿಧೇಯ ಮಗನಾಗಿತ್ತು.

ಪ್ರೀತಿಯ ಸಹೋದರನಾಗಿ ಭಗವಾನ್ ರಾಮ.

ರಾಮಾಯಣದ ಸಂಪೂರ್ಣ ಕಥೆಯಲ್ಲಿ, ಅದರ ನಾಯಕ ರಾಮನು ಹಿರಿಯ ಸಹೋದರನಾಗಿ ಜವಾಬ್ದಾರಿಯಿಂದ ವರ್ತಿಸಿದ್ದಾನೆ ಮತ್ತು ತನ್ನ ಸಹೋದರರಿಗೆ ಸಾಕಷ್ಟು ಪ್ರೀತಿ ಮತ್ತು ಕಾಳಜಿಯನ್ನು ನೀಡಿದ್ದಾನೆ. ಅವನು ಲಕ್ಷ್ಮಣನನ್ನು ತನ್ನ ಸಮಾನ ಎಂದು ಪರಿಗಣಿಸಿದನು ಮತ್ತು ಅವನು ದೊಡ್ಡವನಾಗಿದ್ದರಿಂದ ಅವನ ಮೇಲೆ ಪ್ರಾಬಲ್ಯ ಸಾಧಿಸಲಿಲ್ಲ. ಯುದ್ಧದಲ್ಲಿ ಲಕ್ಷ್ಮಣನು ಗಂಭೀರವಾಗಿ ಗಾಯಗೊಂಡಾಗ ಅವನು ಧ್ವಂಸಗೊಂಡನು. ಲಕ್ಷ್ಮಣನ ಜೀವ ಉಳಿಸಲಾಗಿದೆಯೆ ಎಂದು ಖಚಿತಪಡಿಸಿಕೊಳ್ಳಲು ಎಲ್ಲಾ ಪ್ರಯತ್ನಗಳನ್ನು ಮಾಡಬೇಕೆಂದು ಅವರು ಹನುಮನನ್ನು ಕೇಳಿದರು. ರಾಮಾಯಣದುದ್ದಕ್ಕೂ ಅವರು ಲಕ್ಷ್ಮಣರೊಂದಿಗೆ ಹಂಚಿಕೊಂಡ ಸೌಹಾರ್ದವು ಪ್ರಾಚೀನ ಭಾರತದ ಇತಿಹಾಸದಲ್ಲಿ ಬಲವಾದ ಸಹೋದರತ್ವದ ಪ್ರಬಲ ಸ್ತಂಭವಾಗಿ ಉಳಿದಿದೆ. ಅವನು ಎಂದಿಗೂ ತನ್ನ ಯಾವುದೇ ಸಹೋದರರ ಸಾಧನೆ ಅಥವಾ ಯಶಸ್ಸಿನ ಬಗ್ಗೆ ಅಸೂಯೆ ಪಟ್ಟಿಲ್ಲ ಅಥವಾ ಸ್ವಲ್ಪ ಅಸೂಯೆ ಪಟ್ಟವನಲ್ಲ. ಭರತ ರಾಜನಾದಾಗಲೂ ಅವನು ಅದರ ಬಗ್ಗೆ ಸ್ವಲ್ಪ ತಲೆಕೆಡಿಸಿಕೊಳ್ಳಲಿಲ್ಲ. ಅವನ ಸಹೋದರರೆಲ್ಲರೂ ಅವನನ್ನು ಗೌರವಿಸಲು ಮತ್ತು ಅವನನ್ನು ವಿಸ್ಮಯಗೊಳಿಸುವ ವ್ಯಕ್ತಿಯಾಗಿ ನೋಡಲು ಇದು ಕಾರಣವಾಗಿದೆ.

 

ಪ್ರೀತಿಯ ಗಂಡನಾಗಿ ಭಗವಾನ್ ರಾಮ.
ರಾಮನು ಶಿವನ ಪರೀಕ್ಷೆಯಲ್ಲಿ ಉತ್ತೀರ್ಣನಾಗಿ ಶಾಶ್ವತವಾಗಿ ಸುಂದರವಾದ ಸೀತೆಯನ್ನು ಗೆಲ್ಲಲು ಬಾಣವನ್ನು ಸುಲಭವಾಗಿ ಮುರಿದನು. ತನ್ನೊಂದಿಗೆ ಕಾಡುಗಳಲ್ಲಿ ಸೇರಲು ಸೀತೆಯ ಆಯ್ಕೆಯನ್ನು ಅವನು ಗೌರವಿಸಿದನು, ಮತ್ತು ಅವಳಿಗೆ ಸಮಾನ ಅಧಿಕಾರ ಮತ್ತು ನಿರ್ಧಾರ ತೆಗೆದುಕೊಳ್ಳುವ ಶಕ್ತಿಯನ್ನು ಕೊಟ್ಟನು, ಆದರೆ ಅದೇ ಸಮಯದಲ್ಲಿ ಅವನು ಸೀತಾಳನ್ನು ತನ್ನೊಂದಿಗೆ ಸೇರಲು ತಡೆಯಲು ಪ್ರಯತ್ನಿಸುತ್ತಿದ್ದನು, ಏಕೆಂದರೆ ಸೀತಾ ಸುತ್ತಿದ ಕಾಡಿನಲ್ಲಿ ಜೀವನವನ್ನು ನಡೆಸಬೇಕೆಂದು ಅವನು ಬಯಸಿದನು ಪ್ರತಿ ಹಂತದಲ್ಲೂ ತೊಂದರೆಗಳು ಮತ್ತು ಅನಿಶ್ಚಿತತೆಗಳು. ಅರಮನೆಯಲ್ಲಿ ಅವಳ ಆರಾಮದಾಯಕ ಜೀವನವನ್ನು ತ್ಯಾಗ ಮಾಡಬಾರದೆಂದು ಒತ್ತಾಯಿಸಲು ಅವನು ತನ್ನ ಅತ್ಯುತ್ತಮ ಪ್ರಯತ್ನವನ್ನು ಮಾಡಿದನು, ಆದರೆ ಅಂತಿಮವಾಗಿ ಅವನು ಸೀತಾಳ ನಿರ್ಧಾರವನ್ನು ಗೌರವಿಸುತ್ತಾನೆ, ಅವನು ತನ್ನ ಹೆಂಡತಿಯಾಗಿ ತನ್ನ ಕರ್ತವ್ಯವನ್ನು ಪೂರೈಸುವ ನಿರ್ಧಾರವನ್ನು ಬೆಂಬಲಿಸಿದನು, ಹೀಗಾಗಿ ಅವನು ತನ್ನ ಹೆಂಡತಿಯ ಮೇಲೆ ಪ್ರಾಬಲ್ಯ ಹೊಂದಿಲ್ಲ ಎಂದು ಸಾಬೀತುಪಡಿಸಿದನು ಮತ್ತು ಎಲ್ಲಾ ಹಂತಗಳಲ್ಲಿ ಅವಳ ನಿರ್ಧಾರಗಳನ್ನು ಗೌರವಿಸಿದರು. ಆಧುನಿಕ ದಂಪತಿಗಳಿಗಿಂತ ಭಿನ್ನವಾಗಿ, ಅವರ ವಾದಗಳು ಅಹಂ ಅಥವಾ ಮನೋಭಾವವನ್ನು ಹೊಂದಿಲ್ಲ. ಅವರ ಹೆಂಡತಿಯೊಂದಿಗಿನ ಅವರ ಸಂಭಾಷಣೆಗಳು ಶಾಂತ ಮತ್ತು ಶಾಂತಿಯುತ ಸ್ವರವನ್ನು ಆಧರಿಸಿವೆ.

ಅವರು ಕಾಡುಗಳಲ್ಲಿದ್ದಾಗ, ರಾಮನು ತನ್ನ ಹೆಂಡತಿಯನ್ನು ಬಹಳವಾಗಿ ನೋಡಿಕೊಂಡನು ಮತ್ತು ಅವಳು ಅರ್ಹವಾದ ಎಲ್ಲ ಗಮನವನ್ನು ಕೊಟ್ಟನು ಮತ್ತು ಅವಳ ಆಸೆಗಳನ್ನು ಪೂರೈಸಲು ಎಲ್ಲಾ ಪ್ರಯತ್ನಗಳನ್ನು ಮಾಡಿದನು. ಅವನು ಚಿನ್ನದ ಜಿಂಕೆಗಳನ್ನು ಹುಡುಕಲು ಹೋದನು, ಸೀತೆಯು ಬಯಸಿದನು, ಅದು ರಾವಣನ ಬಲೆ ಎಂದು ತಿಳಿದಿದ್ದರೂ ಸಹ, ಮತ್ತು ಅವನ ಸಹ ರಾಕ್ಷಸರು ಕಾರ್ಯಪ್ರವೃತ್ತರಾಗಿದ್ದಾರೆ.

ಸೀತೆಯನ್ನು ರಾವಣನು ಅಪಹರಿಸಿದ್ದಾನೆಂದು ರಾಮನಿಗೆ ತಿಳಿದ ನಂತರ, ನಾವು ರಾಮನ ಭಾವನಾತ್ಮಕ ಭಾಗವನ್ನು ನೋಡುತ್ತೇವೆ. ಅವನು ಭಾವನೆಗಳಿಂದ ಹೊರಬಂದನು, ಮತ್ತು ಅವನ ಹೆಂಡತಿಯನ್ನು ಅಪಹರಿಸಿದ್ದರಿಂದ ತೀವ್ರವಾಗಿ ಅಸಮಾಧಾನಗೊಂಡನು. ನಂತರ ಅವರು ಈ ಭಾವನೆಗಳನ್ನು ಚಾನಲ್ ಮಾಡಿ ಪರಿಪೂರ್ಣ ವ್ಯಕ್ತಿ ಎಂಬ ಕೌಶಲ್ಯವನ್ನು ಹೊರತಂದರು. ಅವರು ಪಾತ್ರ ಮತ್ತು ಇಚಾಶಕ್ತಿಯ ಶಕ್ತಿಯನ್ನು ತೋರಿಸಿದರು. ಅವನು ತನ್ನಿಂದ ಸಾಧ್ಯವಾದಷ್ಟು ಸಂಪನ್ಮೂಲಗಳನ್ನು ಸಂಗ್ರಹಿಸಲು ಪ್ರಾರಂಭಿಸಿದನು, ಮತ್ತು ದುಷ್ಟ ರಾವಣನನ್ನು ವಶಪಡಿಸಿಕೊಳ್ಳಲು ಮತ್ತು ಸೀತೆಯನ್ನು ಲಂಕಾದಿಂದ ರಕ್ಷಿಸಲು ಬೇಕಾದ ಎಲ್ಲಾ ಸಹಾಯಗಳು. ಸುಗ್ರೀವನು ವಾಗ್ದಾನ ಮಾಡಿದಂತೆ ಸೀತೆಯನ್ನು ಹುಡುಕುವ ಸಹಾಯಕ್ಕೆ ಬದಲಾಗಿ ಅವನು ರಾಜ ವಾಲಿಯನ್ನು ಕೊಲ್ಲಲು ಸುಗ್ರೀವನಿಗೆ ಸಹಾಯ ಮಾಡಿದನು. ಅವನಿಗೆ ಸಹಾಯ ಮಾಡಲು ಸಿದ್ಧವಾಗಿರುವ ವನಾರ್ಸೇನ ಎಂಬ ಕೋತಿಗಳ ಸೈನ್ಯವನ್ನು ಅವನು ಹೀಗೆ ಸಂಗ್ರಹಿಸಿದನು. ಸೀತೆಯನ್ನು ಹುಡುಕಲು ಮತ್ತು ಉಳಿಸಲು ಎಲ್ಲಾ ವಿಸ್ತಾರಗಳಿಗೆ ಹೋಗಲು ಸಿದ್ಧರಿರುವ ಅವರ ಶ್ರೇಷ್ಠ ಭಕ್ತ ಹನುಮನಿಂದಲೂ ಅವರಿಗೆ ಸಹಾಯವಾಯಿತು.  ತನ್ನ ಎಲ್ಲಾ ಪಡೆಗಳನ್ನು ಒಟ್ಟುಗೂಡಿಸಿದ ನಂತರ, ಅವನು ಲಂಕಾ ತಲುಪಿದಾಗ, ಅವನು ದುಷ್ಟ ರಾವಣನನ್ನು ಯಶಸ್ವಿಯಾಗಿ ಜಯಿಸಿದನು. ಸೀತಾ ರಾಜ್ಯಕ್ಕೆ ಮರಳಿದ ನಂತರ, ಅಗ್ನಿ ಪರಿಕ್ಷೆಯನ್ನು, ಬೆಂಕಿಯ ಪರೀಕ್ಷೆಯನ್ನು ಮಾಡಲು, ತನ್ನ ಪರಿಶುದ್ಧತೆಯನ್ನು ಸಾಬೀತುಪಡಿಸಲು ಮತ್ತು ಅಯೋಧ್ಯೆಯಲ್ಲಿನ ಗೊಣಗಾಟಗಳನ್ನು ಒಮ್ಮೆ ಮತ್ತು ಒಮ್ಮೆ ಮುಚ್ಚುವಂತೆ ರಾಮನು ಸೀತಾಗೆ ಆದೇಶಿಸುತ್ತಾನೆ. ಕಾಮವನ್ನು ಹೊಂದಿರುವ ರಾವಣನೊಂದಿಗೆ ತಿಂಗಳುಗಳ ಕಾಲ ಉಳಿದುಕೊಂಡಿದ್ದರೂ ಸಹ, ಸೀತೆಯು ಪರಿಶುದ್ಧನೆಂದು ತಿಳಿದಿದ್ದರಿಂದ, ಪರೀಕ್ಷೆಯನ್ನು ಮಾಡಲು ರಾಮನು ಸೀತಾಗೆ ಆದೇಶಿಸುತ್ತಾನೆ. ಸೀತಾಳ ನಿಷ್ಠೆಯ ಬಗ್ಗೆ ಅವನಿಗೆ ಸಂಪೂರ್ಣ ಮತ್ತು ಬದಲಾಯಿಸಲಾಗದ ನಂಬಿಕೆ ಇತ್ತು, ಅವಳು ಅವಳಿಂದ ದೂರವಾಗಿದ್ದಾಗಲೂ, ಇನ್ನೊಬ್ಬ ಪುರುಷನೊಂದಿಗೆ, ಇನ್ನೊಂದು ಭೂಮಿಯಲ್ಲಿ. ಸೀತಾ ಬೆಂಕಿಯ ಪರೀಕ್ಷೆಯನ್ನು ಯಶಸ್ವಿಯಾಗಿ ಪಾಸಾದಾಗ ಅವನ ನಂಬಿಕೆಯನ್ನು ಪುನರುಚ್ಚರಿಸಲಾಯಿತು. ಇದನ್ನು ಅನುಸರಿಸಿ, ಅವರು ರಾಮನ ಪಟ್ಟಾಭಿಷೇಕಕ್ಕಾಗಿ ರಾಜ್ಯಕ್ಕೆ ಹಿಂತಿರುಗುತ್ತಾರೆ ಮತ್ತು ಅದರ ನಂತರ ಸೀತಾ ಗರ್ಭಿಣಿಯಾಗುತ್ತಾಳೆ.
ವಾಲ್ಮೀಕಿಯ ರಾಮಾಯಣದಲ್ಲಿ ಇದರ ನಂತರ ಏನಾಗುತ್ತದೆ, ಅವನು ಪರಿಪೂರ್ಣ ಗಂಡನಾಗಿದ್ದಾನೆ ಎಂಬ ಪ್ರಶ್ನೆಗಳನ್ನು ಹುಟ್ಟುಹಾಕುತ್ತಾನೆ, ಅವನು ಸೀತೆಯನ್ನು ಗಡಿಪಾರು ಮಾಡಲು ಕಳುಹಿಸಬೇಕಾಗಿತ್ತು, ಏಕೆಂದರೆ ಅವನ ರಾಜ್ಯದ ಜನರು ರಾವಣನೊಂದಿಗೆ ಉಳಿದುಕೊಂಡ ನಂತರವೂ ಸೀತೆಯ ಪರಿಶುದ್ಧತೆಯನ್ನು ನಂಬಲಿಲ್ಲ. ರಾಜನಾಗಿ ತನ್ನ ಕರ್ತವ್ಯ ಮತ್ತು ಗಂಡನಾಗಿ ಅವನ ಕರ್ತವ್ಯದ ನಡುವೆ ಅವನು ಹರಿದುಹೋದನು. ಅವನ ಎಲ್ಲಾ ಪ್ರಜೆಗಳು ಮತ್ತು ಸಾಮ್ರಾಜ್ಯದ ಸಲಹೆಗಾರರು ಸೀತೆಯನ್ನು ಮತ್ತೆ ಗಡಿಪಾರು ಮಾಡಬೇಕೆಂದು ನಂಬಿದ್ದರು. ರಾಜನಾಗಿ ತನ್ನ ಕರ್ತವ್ಯಗಳನ್ನು ನಿರ್ವಹಿಸಲು ಮುಂದಾಗಿದ್ದರಿಂದ ರಾಮನ ರಾಜನಾಗಿ ಸೀತೆಯನ್ನು ಗಡಿಪಾರು ಮಾಡಲು ಸಾಧ್ಯವಾಗಲಿಲ್ಲ. ಆದ್ದರಿಂದ, ಭಾರವಾದ ಹೃದಯದಿಂದ, ಅವನು ಸೀತೆಯನ್ನು ಗಡಿಪಾರು ಮಾಡಲು ಕಳುಹಿಸಬೇಕಾಯಿತು. ಆದಾಗ್ಯೂ, ಸೀತಾ ಅನುಪಸ್ಥಿತಿಯಲ್ಲಿ ಅವನು ಇತರ ಮಹಿಳೆಯರನ್ನು ನೋಡಲಿಲ್ಲ ಅಥವಾ ಮತ್ತೆ ಮದುವೆಯಾಗಲಿಲ್ಲ. ಅವರು ಯಾವಾಗಲೂ ಸೀತಾಗೆ ಶಾಶ್ವತವಾಗಿ ಭಕ್ತಿ ಹೊಂದಿದ್ದರು. ಅವಳೊಂದಿಗೆ ಅವನ ನಿಷ್ಠೆ ಎಂದಿಗೂ ಅಲೆಯಲಿಲ್ಲ. ರಾಮನು ಪರಿಪೂರ್ಣ ಗಂಡ ಮತ್ತು ಪರಿಪೂರ್ಣ ವ್ಯಕ್ತಿಯಾಗಿದ್ದು, ತನ್ನ ರಾಜ್ಯದಲ್ಲಿ ಜನರಿಗೆ ತಮ್ಮ ರಾಜನಾಗಿ ಸೇವೆ ಸಲ್ಲಿಸುವ ತನ್ನ ಬದ್ಧತೆಗೆ ಅಂಟಿಕೊಳ್ಳಬೇಕಾಗಿತ್ತು. ಅವನು ದೈಹಿಕವಾಗಿ ಸೀತೆಯಿಂದ ಬೇರ್ಪಟ್ಟಿದ್ದರೂ, ಅವನ ಆತ್ಮವು ಯಾವಾಗಲೂ ಅವಳೊಂದಿಗೆ ಇತ್ತು. ಅವರು ಶಾಶ್ವತ ಪ್ರೀತಿಯಿಂದ ಸುತ್ತುವರಿಯಲ್ಪಟ್ಟರು ಮತ್ತು ಬೇರ್ಪಟ್ಟ ನಂತರವೂ ಗಂಡ ಮತ್ತು ಹೆಂಡತಿಯಾಗಿದ್ದರು. ಕೊನೆಯವರೆಗೂ, ರಾಮನು ಸೀತೆಗೆ ಸೇರಿದವನು, ಮತ್ತು ಯಾವಾಗಲೂ ಅವಳಿಗೆ ಸೇರುತ್ತಾನೆ.

ರಾಜ ಮತ್ತು ಯೋಧನಾಗಿ ಲಾರ್ಡ್ ರಾಮನ ಸಾಮರ್ಥ್ಯಗಳು
ರಾಮನು ಮಹಾನ್ ಯೋಧ ಮತ್ತು ಉದಾತ್ತ ರಾಜನೆಂದು ಪ್ರಸಿದ್ಧನಾಗಿದ್ದನು. ಅವರು ಅಸಾಧಾರಣ ಗುಣಗಳಿಂದ ಆಶೀರ್ವದಿಸಲ್ಪಟ್ಟರು ಮತ್ತು ಧರ್ಮಗ್ರಂಥಗಳು, ವೇದಗಳು, ಯುದ್ಧಗಳ ಮೇಲೆ ಪಾಂಡಿತ್ಯ ಹೊಂದಿದ್ದರು. ಅವರು ಅತ್ಯುತ್ತಮ ಆಡಳಿತ ಕೌಶಲ್ಯಗಳನ್ನು ಹೊಂದಿದ್ದರು ಮತ್ತು ತುಂಬಾ ಬುದ್ಧಿವಂತರಾಗಿದ್ದರು. ವಿಶ್ವಮಿತ್ರನು ತನ್ನ ತ್ಯಾಗದ ಪ್ರಾರ್ಥನೆಯಲ್ಲಿ ಹಸ್ತಕ್ಷೇಪ ಮಾಡುತ್ತಿದ್ದ ರಾಕ್ಷಸರನ್ನು ಸೋಲಿಸಲು ರಾಮನನ್ನು ಆರಿಸಿದನು. ವಿಶ್ವಮಿತ್ರ ಋಷಿಯನ್ನು ತೊಂದರೆಗೊಳಿಸುತ್ತಿದ್ದ ರಾಕ್ಷಸರನ್ನು ಸೋಲಿಸಲು ರಾಮ, ಲಕ್ಷ್ಮಣರೊಂದಿಗೆ ಆಕಾಶ ಶಸ್ತ್ರಾಸ್ತ್ರಗಳನ್ನು ಬಳಸಿದನು. ಸೀತೆಯನ್ನು ಗೆಲ್ಲಲು ಶಿವನ ಪರೀಕ್ಷೆಯಲ್ಲಿ ಅವನ ಕೌಶಲ್ಯಗಳನ್ನು ಪ್ರದರ್ಶಿಸಲಾಯಿತು, ಅಲ್ಲಿ ಅವನು ತನ್ನ ಪ್ರಬಲವಾದ ಬಿಲ್ಲನ್ನು ಸುಲಭವಾಗಿ ಎತ್ತುತ್ತಾನೆ, ಅದನ್ನು ಎಳೆಯಲು ಸುಮಾರು 5000 ಪುರುಷರು ಬೇಕಾಗಿದ್ದರು.

ರಾವಣ ಮತ್ತು ಅವನ ಸೈನ್ಯದ ವಿರುದ್ಧದ ಮಹಾ ಯುದ್ಧದಲ್ಲಿ, ರಾಮನು ರಾವಣನ ಸಾಮ್ರಾಜ್ಯದ ಅತ್ಯಂತ ಶಕ್ತಿಶಾಲಿ ರಾಕ್ಷಸ ಯೋಧರನ್ನು ಕೊಂದನು, ರಾವಣನ ಸಹೋದರ ಕುಂಭಕರ್ಣ ಮತ್ತು ಸಾವಿರಾರು ರಕ್ಷಾ ಸೈನಿಕರು ಸೇರಿದಂತೆ. ರಾವಣನೊಂದಿಗಿನ ತನ್ನ ಅಂತಿಮ ಮುಖದಲ್ಲಿ, ರಾವಣನು ತನ್ನ ಎಲ್ಲಾ ಮಾಂತ್ರಿಕ ಶಕ್ತಿಗಳನ್ನು, ಅವನ ವರಗಳನ್ನು ಸಹ, ಬ್ರಹ್ಮ ಮತ್ತು ಶಿವನಿಂದ ನೀಡಲ್ಪಟ್ಟನು, ಜೊತೆಗೆ ರಾಕ್ಷಸರು ಹೊಂದಿದ್ದ ಎಲ್ಲಾ ಮಾಂತ್ರಿಕ ಯುದ್ಧಗಳನ್ನೂ ಆಹ್ವಾನಿಸಿದನು. ರಾಮ ಮತ್ತು ರಾವಣನ ನಡುವಿನ ಈ ಯುದ್ಧವು ಪ್ರಾಚೀನ ಹಿಂದೂ ಪುರಾಣಗಳ ಇತಿಹಾಸದಲ್ಲಿ ನಡೆದ ಭೀಕರ ಯುದ್ಧಗಳಲ್ಲಿ ಒಂದಾಗಿದೆ. ಇಬ್ಬರೂ ಒಬ್ಬರಿಗೊಬ್ಬರು ಗಂಭೀರವಾದ ಗಾಯಗಳನ್ನು ಉಂಟುಮಾಡಿದರು, ಏಕೆಂದರೆ ಇಬ್ಬರೂ ಹೆಚ್ಚು ಪ್ರಬಲವಾದ ಶಸ್ತ್ರಾಸ್ತ್ರಗಳನ್ನು ಹೊಂದಿದ್ದರು, ಇಡೀ ವಿಶ್ವವನ್ನು ನಾಶಮಾಡುವ ಸಾಮರ್ಥ್ಯ ಹೊಂದಿದ್ದರು. ಅಂತಿಮವಾಗಿ, ಸುದೀರ್ಘ ಮತ್ತು ದಣಿದ ಯುದ್ಧದ ನಂತರ, ರಾಮನು ರಾವಣನ ಎಲ್ಲಾ ಕೇಂದ್ರಗಳ ಮುಖ್ಯಸ್ಥನಾದ ರಾವಣನ ಕೇಂದ್ರ ಮುಖ್ಯಸ್ಥನನ್ನು ನಾಶಮಾಡಿದನು. ರಾಮ, ತಾನು ಅತ್ಯಂತ ಪ್ರತಿಭಾನ್ವಿತ ಯೋಧರಲ್ಲಿ ಒಬ್ಬನೆಂದು ಸಾಬೀತುಪಡಿಸುತ್ತಾನೆ, ಏಕೆಂದರೆ ಅವನು ಸಾರ್ವಕಾಲಿಕ ಪ್ರಬಲ ಮತ್ತು ಅತ್ಯಂತ ಸಂಪನ್ಮೂಲ ದೆವ್ವಗಳಲ್ಲಿ ಒಬ್ಬನನ್ನು ಸೋಲಿಸುವಲ್ಲಿ ಯಶಸ್ವಿಯಾಗಿದ್ದನು, ಅವನು ತನ್ನ ವರ್ಷಗಳ ಶ್ರದ್ಧೆ ಮತ್ತು ತಪಸ್ಯದಿಂದ ಬ್ರಹ್ಮನಿಗೆ ತನ್ನ ಎಲ್ಲಾ ಶಕ್ತಿಯನ್ನು ಗಳಿಸಿದನು.'''
```
